# 2. exchange battery swap

```python
exchange_battery_swap = pd.read_csv('/content/drive/MyDrive/캡스톤디자인/샘플데이터/exchangeBatterySwap.csv')
#칼럼 선택
columns= [ '_id','exSerial','time','inStTime','inEndTime', 'outEndTime','inASlot','inBSlot','outCSlot','outDSlot']
exchange_battery_swap=exchange_battery_swap[columns]
```

```python
exchange_battery_swap.head(30)
```

```python
print(len(exchange_battery_swap['exSerial'].unique()))
exchange_battery_swap['exSerial'].unique()
```

```python
exchange_battery_swap[exchange_battery_swap['exSerial']=='S2222Z0004'].head(30)
```

```python
#time 형식변환
exchange_battery_swap['real_time'] = exchange_battery_swap['time'].apply(lambda x: time.strftime('%Y/%m/%d %H시 %M분 %S초',time.localtime(x/1000)))
exchange_battery_swap['in_start_time'] = exchange_battery_swap['inStTime'].apply(lambda x: time.strftime('%Y/%m/%d %H시 %M분 %S초',time.localtime(x/1000)))

exchange_battery_swap['in_end_time'] = exchange_battery_swap['inEndTime'].apply(lambda x: time.strftime('%Y/%m/%d %H시 %M분 %S초',time.localtime(x/1000)))

exchange_battery_swap['out_time'] = exchange_battery_swap['outEndTime'].apply(lambda x: time.strftime('%Y/%m/%d %H시 %M분 %S초',time.localtime(x/1000)))
```
